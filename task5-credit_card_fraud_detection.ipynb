Based on th to the type of transaction - Fraud or Legit, Separating the Dataset

```python
legit = df[df.Class == 0]
```

```python
fraud = df[df.Class==1]
```

Difference persist in the mean value for our normal transaction and mean value for ur fraud transactions & to balance the data for legit and fraud transaction value points we will use sampling for creating a new dataset of normal transactions with 492 entries being selected randomly out of 284315

```python
normal_sample = legit.sample(n=492)
```

Merging the two datasets for fraud and legit transactions with equal number of sample points

```python
new_dataset = pd.concat([normal_sample, fraud], axis = 0) # axis =0 species row wise joining of the datasets
```

```python
new_dataset
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean() 
```

The time feature can be dropped and using a derived column with time delta function to represent the duration (difference between two time values)

```python
delta_time = pd.to_timedelta(new_dataset['Time'], unit = 's')
# create the derived column
new_dataset['time_hour']=(delta_time.dt.components.hours).astype(int)
# now drop the time column
new_dataset.drop(columns='Time', axis=1, inplace = True)
```

```python
new_dataset
```

Separating the features and target variables

```python
x = new_dataset.drop('Class', axis=1)
```

```python
y = new_dataset['Class']
```

```python
x.shape
```

```python
y.shape
```

Splitting the data into training and testing data 

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.25, random_state = 3, stratify = y)
```

Accumulating all the column names under one variable

```python
cols = list(x.columns.values)
```

```python
normal_entries = new_dataset.Class==0
fraud_entries = new_dataset.Class==1

plt.figure(figsize=(20,70))
for n, col in enumerate(cols):
    plt.subplot(10,3,n+1)
    sns.histplot(x[col][normal_entries], color='blue', kde = True, stat = 'density')
    sns.histplot(x[col][fraud_entries], color='red', kde = True, stat = 'density')
    plt.title(col, fontsize=17)
plt.show()
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_train)
pred_test = model.predict(x_test)
```

# Model evaluation

```python
# creating confusion matrix
from sklearn.metrics import confusion_matrix
def Plot_confusion_matrix(y_test, y_pred):
    cm = confusion_matrix(y_test,pred_test)
    plt.clf()
    plt.show()
```

```python
# accuracy on training data
acc_score= round(accuracy_score(y_pred, y_train)*100,2)
```

```python
print('the accuracy score for training data of our model is :', acc_score)
```

```python
y_pred = model.predict(x_test)
acc_score = round(accuracy_score(y_pred, y_test)*100,2)
```
